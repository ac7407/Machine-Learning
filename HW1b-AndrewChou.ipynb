```python
# import modules
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# download and read data from excel file
data_root = "/home/jovyan/delaney-processed.csv"
delaney = pd.read_csv(data_root)

# visualize data of solubility vs. molecular weight
delaney.plot(kind='scatter', grid=True,
                x="Molecular Weight", y="measured log solubility in mols per litre")
plt.axis([0, 800, -10, 2])
plt.title("Solubility vs. Molecular Weight")
plt.show()
```

```python
# create linear regression model
X = delaney["Molecular Weight"].values.reshape(-1, 1)
y = delaney["measured log solubility in mols per litre"].values.reshape(-1, 1)
model = LinearRegression()
model.fit(X, y)

# visualize linear regression model
plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X), color='red')
plt.title("Linear Regression Model of Solubility vs. Molecular Weight")
plt.xlabel("Molecular Weight")
plt.ylabel("measured log solubility in mols per litre")
plt.show()

# predict solubility of a molecule with molecular weight of 500
print("Predicted solubility of a molecule with molecular weight of 500: ", model.predict([[500]]))
```
